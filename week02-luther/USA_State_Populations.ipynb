```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
state_pops = pd.read_csv('USA_Population_of_States_US_Census_Intercensal_Tables_1917-2017.csv')
state_pops.sample(10)
```

```python
states = ['CA','IL','TX','NY']
for state in states:
    pops = state_pops[state]
    plt.plot(state_pops.Year, pops, label=state)
plt.legend();
```
